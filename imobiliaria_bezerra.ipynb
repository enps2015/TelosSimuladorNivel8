```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dados
df = pd.read_excel('Imobiliaria_Bezerra.xlsx')  # ajuste o nome do arquivo conforme necessário

# Análise por vendedor
analise_vendedores = df.groupby('Vendedor').agg({
    'Valor de Compra': ['count', 'sum', 'mean'],
    'Comissão do Vendedor': 'sum'
}).round(2)

# Renomear colunas
analise_vendedores.columns = ['Qtd_Vendas', 'Valor_Total_Vendas', 'Valor_Medio_Vendas', 'Comissao_Total']

# Ordenar por valor total de vendas
top_vendedores = analise_vendedores.sort_values('Valor_Total_Vendas', ascending=False).head(3)

# Visualização
plt.figure(figsize=(12, 6))
sns.barplot(data=top_vendedores.reset_index(), x='Vendedor', y='Valor_Total_Vendas')
plt.title('Top 3 Vendedores por Valor Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```
